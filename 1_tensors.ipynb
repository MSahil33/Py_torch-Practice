### What are Tensors ?
Tensors is a specialized multi-dimensional array designed for mathematical and computational efficiency.

***Tensors can be in any dimensions :***
1. Scalar => 0-dimensional tensors(a single number)

    - Ex: 5.0 , -3.14...etc
2. Vectors => 1-dimensional tensors(a list of numbers)

    - Ex: [0.12,3.1,21.34,21.45]
3. Matrices => 2-dimensional tensors(a 2D grid of numbers)

    - Ex : [[1.2,3.0,244]
         [8.0,2.1,3.4]]

4. 3D tensors (Coloured image)

    - Ex : RGB image (256,256,3) -> Here 256x256 represents a 2d tensors which is in 3 dimensions(3 colors i,e. R,G,B)

5. 4D tensors (Batches of RGB images = video)

    - Ex : (32,128,128,3) A batch of 32 images, each of size 128x128 with 3 colour channels(RGB)

6. 5D tensors (Video data = Group of videos)

    - Ex : (10,16,64,64,3) A batch of 10 video clips, each with 16 frames of size 64x64 and 3 channels (RGB)

#### Why are Tensors Useful ?

***1. Mathematical Operations :***

Tensors enable efficient mathematical computations necessary for neural network operations.

***2. Representation of Real-world Data :***

Data like images, audio, Video and Text can be easily represented as tensors.

***3. Efficient Computations :***

Tensors are optimized for hardware acceleration , allowing parallel computations on GPUs or TPUs, which are crucial for training deep learning models

#### In-Place Operations

To do the in-place operations on the tensors we have to just append the underscore "_" at the end of the operation we have to perform.

Ex : x.add_(y), x.relu_(y)

```python
tr_j = torch.randint(size=(2,3),low=0,high=10)
tr_k = torch.randint(size=(2,3),low=0,high=10)
print(tr_j)
print(tr_k)
```

```python
# Addition In-place operation
tr_j.add_(tr_k) # Here change will be applicable to original tr_j tensor
```

```python
tr_j
```

```python
tr_l = torch.rand(2,3,4)
print("Before Permute ",tr_l.shape)

# Permute operation : we can rearrange the dimesnions of the tensor with permute
tr_l = tr_l.permute(2,0,1) # Here 2,0,1 : Indices => (2,3,4): dimensions means the position of the dimensions is rearranged i,e 2 is replace with 4 dimension, 3 with 2 and 4 with 3 dimesnions
# So now shape will transform from (2,3,4) to (4,2,3)
print("After permute : ",tr_l.shape)
```
