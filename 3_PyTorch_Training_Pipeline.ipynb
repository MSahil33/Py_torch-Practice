#### 3. Model Training Process

***Important Parameters***

```python
learning_rate = 0.1
epochs = 50 # Training loop count
```

```python
X_train.shape
```

##### 1. Defining Model

```python
class MySimpleNN:
  def __init__(self,X_train):
    # Defining random weights of shape (30,1) ; Here 30 is the num of input features
    self.weights = torch.rand(size=(X_train.shape[1],1),dtype=torch.float64,requires_grad=True)
    # Defining bias
    self.bias = torch.zeros(1,dtype=torch.float64,requires_grad=True)

  # 2. Forward Propogation
  def forward_prop(self,X_train):
    """
      Formula :
        z = weights * X + bias
        ypred = sigmoid(z)
    """
    z = torch.matmul(X_train,self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  # 3. Loss Calculation
  def loss_calculation(self,y_pred,y_train):
    """
      Binary cross entropy loss function
      Formula :
        L =  -[y_train * ln(y_pred) + (1-y_train) * ln(1 - y_pred)]
    """

    epsilon = 1e-7
    # Clipping or clamping y_pred values
    y_pred = torch.clamp(y_pred,min=epsilon,max=1 - epsilon)

    loss = - (y_train * torch.log(y_pred) + (1-y_train) * torch.log(1-y_pred)).mean()

    return loss
```

```python
##### 1. Defining the model
model = MySimpleNN(X_train_tr)
```

##### Training Process for the num of epochs

```python
for epoch in range(epochs):

  ##### 2. Forward Propogation
  y_pred = model.forward_prop(X_train_tr)

  ##### 3. Loss Calculation
  loss = model.loss_calculation(y_pred,y_train_tr)
  # print("Loss : ",loss)

  ##### 4. Backward Propogation
  loss.backward() # Using Autograd

  ##### 5. Parameters Updates
  """
    Formula to update parameters:
      Wnew = Wold - learning_rate * (dL/dW)
      Bnew = Bold - learning_rate * (dL/dB)
  """

  # Not caluculatin gradient for the updated weights and bias
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  # Cleaning gradients so that it should accumulate for the next run
  model.weights.grad.zero_()
  model.bias.grad.zero_()


  print(f"Epoch : {epoch + 1} , Loss : {loss:.4f}")
```

#### 4. Model Evaluation

```python
with torch.no_grad():
  y_pred = model.forward_prop(X_test_tr)

  # As we are getting the y_pred in the range of 0 to 1, but we want binary 0 or 1 so we are normalizing it
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_test_tr).float().mean()

print(f"Accuracy : {accuracy.item():.4f}")
```
